### What is intelligence?

A crude, yet still useful definition of intelligence is:  

* ***intelligence:*** *doing the right thing.*

The above deifintion depends highly upon context. Widely regarded as the father of Theoretical Computer Science and AI, **Alan Turing** provided an experimental defintion of when a system has acheived artificial intelligence in 1950 known as the **Imitation Game** (and later more commonly refered to as the **Turing Test**). More specifically, in an academic paper entitled "Computing Machinery and Intelligence" in the Journal Mind, Alan Turing asked "Can machines think?" 
[(see the paper here!)](https://courses.cs.umbc.edu/471/papers/turing.pdf) Considering that it is difficult, if not "absurd" to define what it means to think, Turing instead framed the question as to *whether a machine can **act** like a human as observed by a human **interrogator**/judge as compared to another human (respondant).  

* **Imitation Game (aka Turing Test)** - Machine A, and human Respondant B, are questioned by human Interrogator C. The Machine and Respondant are both trying to convice the Interrogator that they are human.

### Machine Learning

***automatically learning from data to understand and predict***

The three main areas of Machine Learning are:  

1. ***supervised learning*** - in which input data has defined desired outputs or **labels** which we wish for a model to learn. Trained models can then be used to predict outputs for unseen input data.
1. ***unsupervised learning*** - this involves when we have data with no defined output labels , our desire is to automatically identify patterns in the data to help us undertand. *Dimensionality reduction* and *clustering* data into meaningful groups are two common techniques.
1. ***reinforcement learning*** - training an agent to identify appropriate actions through interacting with an environment or based on past interactions and outcomes. This is how more modern chess or other game programs learn how to play (eg. AlphaGo, AlphaZero).

### The Representation Problem

This refers to the challenge of representing a real world problem in a computing system, often in a much simpler format. Often a real world problem can be represented in many different ways. The way which is chosen can heavily influence our success and efficiency in solving the given problem.

### PEAS AI System Model

The PEAS model is often a useful way to provide a simple view of a real world problem for which we wish to develop some form of AI for.


**PEAS Agent Loop** - a way of modeling a system in which an AI agent can act in an environment in order to optimize some performance measure. An agent senses the environment's state, thinks, chooses an action, and repeats.
1. **Performance measure** - the metric which defines desired outcomes and indirectly intelligent behavior. 
1. **Environment** - defines a system in terms of a **state** and how actions change the state.
1. **Actions** - All the options that an agent can take.
1. **Sensors** - Signals that the agent receives providing a view of the system state.

Example: Tic Tac Toe

An AI agent which plays Tic Tac Toe against a human opponent. The game **state** can be defined as the cells the X and O have already played in. The **actions** can be the valid moves the agent can currently take. The **sensors** can be the full game board view (contrast this to games where there is a *fog of war* and agents can only see what is close to them in a 3d or 2d world). The **performance metric** for an agent can be the number of wins - number of losses after playing 100 games against an opponent.
